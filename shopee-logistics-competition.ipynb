```python
import pandas as pd

# load the data
data = pd.read_csv("/kaggle/input/delivery-orders/delivery_orders_march.csv")

# check the first 5 rows
data.head()
```

```python
# get the last words

# buyeraddress
u = data['buyeraddress'].str.partition()
Destination = pd.DataFrame({'destination': u[2].str.split().str[-1]})
data["destination"] = Destination
data["destination"] = data["destination"].str.lower()

# selleraddress
u = data['selleraddress'].str.partition()
Origin = pd.DataFrame({'origin': u[2].str.split().str[-1]})
data["origin"] = Origin
data["origin"] = data["origin"].str.lower()

# check the first 5 rows
data.head()
```

```python
# get SLA Matrix

def sla(data):
    if data["destination"] == "manila" and data["origin"] == "manila":
        return 3
    elif data["destination"] == "manila" and data["origin"] == "luzon":
        return 5
    elif data["destination"] == "manila" and data["origin"] == "visayas":
        return 7
    elif data["destination"] == "manila" and data["origin"] == "mindanao":
        return 7
    elif data["destination"] == "luzon" and data["origin"] == "manila":
        return 5
    elif data["destination"] == "luzon" and data["origin"] == "luzon":
        return 5
    elif data["destination"] == "luzon" and data["origin"] == "visayas":
        return 7
    elif data["destination"] == "luzon" and data["origin"] == "mindanao":
        return 7
    elif data["destination"] == "visayas" and data["origin"] == "manila":
        return 7
    elif data["destination"] == "visayas" and data["origin"] == "luzon":
        return 7
    elif data["destination"] == "visayas" and data["origin"] == "visayas":
        return 7
    elif data["destination"] == "visayas" and data["origin"] == "mindanao":
        return 7
    elif data["destination"] == "mindanao" and data["origin"] == "manila":
        return 7
    elif data["destination"] == "mindanao" and data["origin"] == "luzon":
        return 7
    elif data["destination"] == "mindanao" and data["origin"] == "visayas":
        return 7
    else:
        return 7
    


data["sla"] = data.apply(sla,axis=1)

# check the first 5 rows
data.head()
```

```python
# Initiate public holidays
public_holidays = ["2020-03-08", "2020-03-25", "2020-03-30", "2020-03-31"]
```

```python
# convert epoch time to standard datetime
data['pickup_time'] = pd.to_datetime(data['pick'], unit = 's').dt.date
data['1st_deliver_time'] = pd.to_datetime(data['1st_deliver_attempt'], unit = 's').dt.date
data['2nd_deliver_time'] = pd.to_datetime(data['2nd_deliver_attempt'], unit = 's').dt.date

# check the first 5 rows
data.head()
```

```python
from datetime import datetime

# calculate the days taken by the logistics provider for 1st delivery attempt (excluding Sundays, and Public Holidays)
def get_busday_first(data):
    create_date = str(data['pickup_time'])
    resolve_date = str(data['1st_deliver_time'])

    create_datetime = datetime.strptime(create_date, '%Y-%m-%d')
    resolve_datetime = datetime.strptime(resolve_date, '%Y-%m-%d')

    busday = np.busday_count(create_date, resolve_date, holidays=public_holidays, weekmask=[1,1,1,1,1,1,0])

    return busday  
```

```python
data['1st_deliver_days'] = data.apply(get_busday_first, axis=1)

# check the first 5 rows
data.head()
```

```python
# calculate the days taken by the logistics provider for 2nd delivery attempt (excluding Sundays, and Public Holidays)
def get_busday_second(data):
    create_date = str(data['1st_deliver_time'])
    resolve_date = str(data['2nd_deliver_time'])
    
    if resolve_date == 'NaT':
        resolve_date = create_date

    create_datetime = datetime.strptime(create_date, '%Y-%m-%d')
    resolve_datetime = datetime.strptime(resolve_date, '%Y-%m-%d')

    busday = np.busday_count(create_date, resolve_date, holidays=public_holidays, weekmask=[1,1,1,1,1,1,0])

    return busday   
```

```python
data['2nd_deliver_days'] = data.apply(get_busday_second, axis=1)

# check the first 5 rows
data.head()
```

```python
# determine if the first delivery is late or not late
def latefirst(data):
    if data['1st_deliver_days'] <= data['sla']: # first delivery not late
        return 0
    else:
        return 1
```

```python
data["is_late_1"] = data.apply(latefirst,axis=1)

# check the first 5 rows
data.head()
```

```python
# determine if the second delivery is late or not late
def latesecond(data):
    if data['2nd_deliver_days'] <= 3: # second delivery not late
        return 0
    else:
        return 1
```

```python
data["is_late_2"] = data.apply(latesecond,axis=1)

# check the first 5 rows
data.head()
```

```python
# sum the values, so even if the first delivery is late and the second delivery is on time, it's still counted as late
not_late = data[(data["is_late_1"] + data["is_late_2"] == 0)][['orderid']]
not_late['is_late'] = 0
not_late.head()
```

```python
late = data[~data['orderid'].isin(not_late['orderid'])][['orderid']]
late['is_late'] = 1
late.head()
```

```python
final = pd.concat([late, not_late], axis=0)
final = final.sort_index(axis = 0) 
final.head()
```

```python
final.to_csv("solution.csv", index=False)
```
